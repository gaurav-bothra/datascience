```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
j = cv2.imread('j.jpg')
plt.imshow(j)
```

```python
kernel = np.ones((5,5), np.uint8)
```

```python
erosion = cv2.erode(j,kernel,iterations = 1)
```

```python
plt.imshow(erosion)
```

```python
dilation = cv2.dilate(erosion,kernel,iterations = 1)
```

```python
plt.imshow(dilation)
```
